## Rotary encoders provide angular position measurements in "counts"


The Romi's wheel motors are each equipped with a rotary encoder. The conversion from counts to radians depends on the gear ratio, counts per revolution, etc. and can be computed using dimensional analysis. Assuming a 1:1 gear ratio, an example is provided below.

\begin{equation}
\require{cancel}
rad = \frac{1 \cancel{rev}}{CPR \cancel{count}}\times\frac{2\pi \:rad}{\cancel{rev}}
\end{equation}

## Rotary potentiometers measure the position of each of Romi's servos.


A rotary potentiometer is simply an adjustable voltage divider!



## Rotary potentiometers measure the position of each of Romi's servos.


\begin{equation}V_{out} = V_{in}\frac{R_2}{R_1+R_2}\end{equation}

## Rotary potentiometers measure the position of each of Romi's servos.


An [analog to digital converted](https://en.wikipedia.org/wiki/Analog-to-digital_converter) converts voltages between 0 and 5V on an Arduino "analog pin" to a 10-bit number, meaning that the relationship between ADC "counts" and "volts" can be derived analytically:

```python
from numpy import *
from matplotlib.pyplot import *
volts = linspace(0,5,100)
counts = volts*1023.0/5.0
figure()
plot(counts,volts,'k')
grid('minor')
ylabel('Voltage')
xlabel('Counts')
```

\begin{equation}
\require{cancel}
V = \cancel{counts} \times \frac{5.0V}{1023\cancel{counts}}
\end{equation}

However, this doesn't give the conversion between volts at the potentiometer wiper and servo arm degrees!

## Sometimes conversions are easier with experimental data
Converting counts to volts on the servos' rotary potentiometers doesn't help convert to position! If I move the servo (by hand, for example) to two known positions, I can produce a best-fit line using point-slope form. Let's say I let a servo achieve steady-state at 90 degrees, and 110 degrees, and then read the potentiometer value in counts.

```python
counts_exp = [308,350]
deg_exp = [90,110]
m = (deg_exp[-1]-deg_exp[0])/(counts_exp[-1]-counts_exp[0])
counts_lin = linspace(0,1023,25)
deg_lin = m*(counts_lin-counts_exp[0])+deg_exp[0]
figure()
plot(counts_exp,deg_exp,'ko',markersize=10)
plot(counts_lin,deg_lin,'r')
legend(['measured','linear fit: m='+str(m)+' deg/count'])
xlabel('ADC counts')
ylabel('degrees')
grid('on')
```

After finding the slope between your two known points, can then find the degree value for any count value using that slope and one of your known combinations of degrees and counts:
\begin{equation}
deg = m(counts-counts_{known})+deg_{known}
\end{equation}

This is *not as accurate* as a true "linear fit" because it does not account for noise! If you wanted to use a true linear fit, you could consult [this section](https://alexanderallenbrown.github.io/ES103_SP23_Students/ES103_Summary_Resource/ES103_Summary_Resource.html#Empirical-Models:-fitting-polynomials-in-MATLAB/Octave) from ES103.

## IR distance sensors are highly nonlinear!

the reading given by the sensor is often very sensitive to IR reflectivity of the environment. 


Sometimes, producing a [lookup table](https://alexanderallenbrown.github.io/ES103_SP23_Students/ES103_Summary_Resource/ES103_Summary_Resource.html#Linear-Interpolation) and using linear interpolation (either using numpy.interp or manually with point-slope form) is the best way to convert from IR "counts" to meters.

## IR distance sensors are highly nonlinear!

```python
ir_data=array([[0.0254,991,.005,],[0.0508,987,.005,],[0.0762,980,.005,],[0.1016,853,.005,],[0.127,712,.005,],[0.1524,585,.005,],[0.1778,502,.005,],[0.2032,435,.005,],[0.2286,383,.005,],[0.254,346,.005,],[0.2794,316,.005,],[0.3048,296,.005,],[0.3302,281,.005,],[0.3556,268,.005,],[0.381,255,.005,],[0.4064,248,.005,],[0.4318,239,.005,],[0.4572,233,.005,],[0.4826,229,.005,],[0.508,223,.005,],[0.5334,219,.005,],[0.5588,215,.005,],[0.5842,211,.005,],[0.6096,209,.005,],[0.635,207,.005,],[0.6604,205,.005,],[0.6858,203,.005,],[0.7112,201,.005,],[0.7366,200,.005]])
ind = 5
m= (ir_data[ind,0]-ir_data[ind+1,0])/(ir_data[ind,1]-ir_data[ind+1,1])
query = int(mean([ir_data[ind,1],ir_data[ind+1,1]]))
interped = m*(query-ir_data[ind,1])+ir_data[ind,0]

xlin = ir_data[ind-5:ind+5,1]
lin = m*(xlin-ir_data[ind,1])+ir_data[ind,0]

figure()
plot(ir_data[:,1],ir_data[:,0],'k.')
plot(xlin,lin,'r')
plot(query,interped,'ro')

legend(['data','interpolated line','interpolated result'])
xlabel('ADC counts')
ylabel('Distance to white paper (m)')
```
