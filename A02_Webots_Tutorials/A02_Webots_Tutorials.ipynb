# Assignment 2: Webots Tutorials

**Due Wednesday, September 4, 2022 by 5pm**

In ES302, we will be modeling and testing robots in simulation before running experiments. This will allow us to predict how our robots will behave before we subject them or their environments to undue risk with an experiment. As we did in ES103, we will use a disciplined process that is rooted in the model->predict->validate paradigm.


As we did in ES103, we will sometimes be employing physics-based differential equation models, lookup table models, or other methods to build simulations of robot behavior "from scratch," but as the complexity of a robot, a group of robots, or a robot-environment system increases, the methods we learned in ES103 will quickly meet their limitations.

To deal with this, we will be using a commercial robot simulation package called [WeBots](www.cyberbotics.com). This software is very powerful, but the development team has created a series of tutorials that make learning how it works quick and easy.
