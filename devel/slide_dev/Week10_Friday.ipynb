## A07 tips

Three ways to measure position and orientation (yaw) of Romi:
1. Ground truth from supervisor
2. Using **measurements from** the encoders to get $\omega_l,\omega_r$, then using Odometry
3. Using **A06 model** to map **input voltages (in counts -> volts)** to get *predictions* of $\omega_r,\omega_l$, and **then** using those predicted angular velocities to perform Odometry

A07 asks you to compare all three of these methods! Remember that your model predicts forward speed $U$ and angular velocity $\dot{\psi}$ directly. But then, we know that:

$$\dot{\psi} = \frac{\omega_rR_{wheel} - \omega_lR_{wheel}}{2R_{romi}}$$

$$U = R_{wheel}\frac{\omega_l+\omega_r}{2}$$

This means that given a predicted $U$ and a predicted $\dot{\psi}$ at any moment in time, you can compute $\omega_l$ and $\omega_r$. **The assignment suggests recording input voltages and time in one file, and raw encoder values in another.** This means you'd do all predictions and odometry calcs "offline" after the simulation is over... but this is your choice.
