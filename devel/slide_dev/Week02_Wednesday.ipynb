## Returning to our discussion of FSMs from Monday:

Once we decide on a set of states for an FSM design, and we draw a state transition diagram, what is our next step?

### State Transition Table
A state transition table allows us to codify the logic represented in a state transition diagram


| **Transition** | **Start** | **Condition** | **End** |
|----------------|-----------|---------------|---------|
|        A        |    state 1       |    (insert logic)           |  state 1       |
|         B       |     state 1      |        (insert logic)       |   state 2      |
|         C       |     state 2      |      (insert logic)         |    state 2     |

| **Transition** | **Start** | **Condition** | **End** |
|----------------|-----------|---------------|---------|
|        A        |    state 1       |    (insert logic)           |  state 1       |
|         B       |     state 1      |        (insert logic)       |   state 2      |
|         C       |     state 2      |      (insert logic)         |    state 2     |

The key thing to note about a transition is that it includes **context** as well as a logical **stimulus**. When "reading" the table, we read "for transition A to be true, the machine must be in state 1 **and** the condition must be true."

## How do we *Implement* an FSM?


Using prof. Brown's "four block structure" makes it very straightforward to turn your state transition diagram into a functioning FSM implementation. The structure is given below.

### Block 1: Process User Inputs & Sensors

Block 1 is also where the program needs to take any "raw" inputs from the user and process/translate them for use in the state transition logic. For example, if your design requires you to hold a button for 2 seconds, you may need to write some logic that uses a "timer" (which we will cover later) to check how long a certain momentary button has been held.



#### Block 1 Example - Unique Press

A common task that needs to go into this portion of your program is that of detecting a *unique press* on a momentary switch or button. Look at the time history of a momentary button press below:

Implementing the "unique button press" logic means capturing the *context* that the button is true now, but was false last time I looked.
```javascript
BTN_PRESS = Button_State && !OLD_Button_State;
```

In Python code, this might look like:

```python
BTN_PRESS = Button_State and not(OLD_Button_State)
```


Of course, this is just one example of what "Block 1" of your program should contain. In summary, it needs to update the value of anything that will be used to evaluate the state transition logic in the next block.
